```python
import pandas as pd

df = pd.read_parquet("hf://datasets/bjoernp/tagesschau-2018-2023/data/train-00000-of-00001-1c2a165f0626c4a6.parquet")
```

```python
df.head()
```

```python
df["count"] = df["link"].apply(lambda x: len(x.split("/")))
```

```python
df.head()
```

```python
# filter out all rows where the count is not equal to 4
df = df[df["count"] == 4]
```

```python
df
```

```python
df["main_tag"] = df["link"].apply(lambda x: x.split("/")[1])
df["sub_tag"] = df["link"].apply(lambda x: x.split("/")[2])
```

```python
df.head()
```

```python
# print unique main tags
df["main_tag"].unique()
```

```python
# group by main tag and count the number of rows
df.groupby("main_tag").count()
```

```python
# group by main tag and print unique sub tags per main tag
df.groupby("main_tag")["sub_tag"].unique()
```

```python
# group by main tag and sub tag and count the number of rows
df.groupby(["main_tag", "sub_tag"]).count()
```

```python
# main tags to remove
main_tags_to_remove = ["multimedia", "investigativ", "europawahl"]
sub_tags_to_remove = ["korrespondenten", "uswahl", "btw21", "regional", "wahlen"]

# filter out all rows where the main tag is in the list of main tags to remove
df = df[~df["main_tag"].isin(main_tags_to_remove)]

# filter out all rows where the sub tag is in the list of sub tags to remove
df = df[~df["sub_tag"].isin(sub_tags_to_remove)]
```

```python
# group by main tag and sub tag and count the number of rows, but only show the count, not the other columns
df.groupby(["main_tag", "sub_tag"]).size()
```

```python
df.head()
```

```python
# remove count column
df = df.drop(columns=["count"])
```

```python
# combine main tag and sub tag
df["tag"] = df["main_tag"] + "/" + df["sub_tag"]
```

```python
# store the cleaned data in a new parquet file
df.to_parquet("tagesschau_cleaned.parquet")
```
